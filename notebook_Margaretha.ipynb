### Cleaning Data

````python
df_all_stations = pd.concat(all_df_stations, ignore_index=True)
df_all_stations = df_all_stations.drop('No', axis=1)
df_all_stations
````

Tangani missing value dengan cara ganti menggunakan nilai median (untuk data numerik) dan nilai modus (untuk data categorical) tiap station
- Nilai median diisi berdasarkan tiap stasiun karena setiap stasiun memiliki karakter unik dan menjaga distribusi asli masing-masing stasiun
- Median digunakan karena sangat toleran terhadap outlier

````python
# Tangani nilai NaN untuk kolom dengan tipe data numerik yaitu
# PM2.5, PM10, SO2, NO2, CO, O3, TEMP, PRES, DEWP, RAIN, dan WSPM
columns_to_fill = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

df_all_stations[columns_to_fill] = (
    df_all_stations
      .groupby('station')[columns_to_fill]
      .transform(lambda col: col.fillna(col.median()))
)
````

````python
# Tangani nilai NaN untuk kolom tipe string menggunakan nilai modus
def fill_mode(series):
  mode_val = series.mode()
  if not mode_val.empty:
    return series.fillna(mode_val.iloc[0])
  else:
    return series.fillna('UNKNOWN')

df_all_stations['wd'] = df_all_stations.groupby('station')['wd'].transform(fill_mode)
````

````python
# Cek semua lagi

# .info
print('Hasil df_all_stations.info()')
print(df_all_stations.info())

# .value_counts
print('Hasil nilai unik tiap kolom')
for column in df_all_stations.iloc[:, 4:]:
  print(df_all_stations[column].value_counts())
````

````python
# Cek outlier setiap kolom
# Kolom numerik seperti PM2.5, PM10, SO2, NO2, CO, O3, TEMP, PRES, DEWP, RAIN, WSPM
# Yang nilai maximumnya sangat jauh dari nilai Q3
columns_to_check_outlier = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

rows = 6
cols_per_row = 2

fig, axes = plt.subplots(rows, cols_per_row, figsize=(14, 20))
axes = axes.flatten()

for i, col in enumerate(columns_to_check_outlier):
  sns.boxplot(data=df_all_stations, x=col, ax=axes[i], color='skyblue')
  axes[i].set_title(f'Boxplot untuk kolom{col}', fontsize=10)
  axes[i].set_xlabel('')
  axes[i].set_ylabel('')

# Nonaktifkan subplot kosong
for j in range(len(columns_to_check_outlier), len(axes)):
  fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
````
